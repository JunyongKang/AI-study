```python
import random

def simulate_dice_rolls(num_rolls):
    total = 0
    for _ in range(num_rolls):
        roll = random.randint(1, 6)  # Simulate a dice roll (1 to 6)
        total += roll
    average = total / num_rolls
    return average

expected_value = (1 + 2 + 3 + 4 + 5 + 6) / 6  # Expected value of a fair 6-sided die

num_simulations = 1000  # Number of simulations

for i in range(1, num_simulations + 1):
    simulated_average = simulate_dice_rolls(i)
    print(f"Simulation {i}: Simulated Average = {simulated_average:.4f}, Expected Value = {expected_value}")
```

```python
import random
import matplotlib.pyplot as plt

def simulate_gaussian_samples(num_samples, mean, std_dev):
    total = 0
    samples = []
    for _ in range(num_samples):
        sample = random.gauss(mean, std_dev)  # Simulate a Gaussian sample
        total += sample
        samples.append(sample)
    average = total / num_samples
    return average, samples

true_mean = 0.0  # True mean of the Gaussian distribution
true_std_dev = 1.0  # True standard deviation of the Gaussian distribution

num_samples = 1000  # Number of samples

simulated_averages = []
for i in range(1, num_samples + 1):
    simulated_average, _ = simulate_gaussian_samples(i, true_mean, true_std_dev)
    simulated_averages.append(simulated_average)

# Plot the convergence of simulated averages to the true mean
plt.plot(range(1, num_samples + 1 - 10), simulated_averages[10:], label='Simulated Averages')
plt.axhline(y=true_mean, color='r', linestyle='--', label='True Mean')
plt.xlabel('Number of Samples')
plt.ylabel('Sampled Average')
plt.title('Law of Large Numbers with Gaussian Distribution')
plt.legend()
plt.show()
```
