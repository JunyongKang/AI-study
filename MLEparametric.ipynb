```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, bernoulli, expon

# Gaussian distribution example
np.random.seed(42)
true_mean_gaussian = 5.0
true_std_gaussian = 2.0
num_samples_gaussian = 100
samples_gaussian = np.random.normal(true_mean_gaussian, true_std_gaussian, num_samples_gaussian)

# MLE solution for mean and standard deviation of a Gaussian distribution
mu_mle_gaussian = np.mean(samples_gaussian)
sigma_mle_gaussian = np.std(samples_gaussian)

# Bernoulli distribution example
true_prob_bernoulli = 0.7
num_samples_bernoulli = 100
samples_bernoulli = bernoulli.rvs(true_prob_bernoulli, size=num_samples_bernoulli)
print(samples_bernoulli)

# MLE solution for the probability of success in a Bernoulli distribution
prob_mle_bernoulli = np.mean(samples_bernoulli)

# Exponential distribution example
true_lambda_exponential = 0.5
num_samples_exponential = 100
samples_exponential = np.random.exponential(scale=1/true_lambda_exponential, size=num_samples_exponential)

# MLE solution for the rate parameter (lambda) of an Exponential distribution
lambda_mle_exponential = 1 / np.mean(samples_exponential)

# Print the true and estimated parameters for all distributions
print("Gaussian Distribution:")
print(f"True Mean: {true_mean_gaussian}, Estimated Mean (MLE): {mu_mle_gaussian}")
print(f"True Std Dev: {true_std_gaussian}, Estimated Std Dev (MLE): {sigma_mle_gaussian}\n")

print("Bernoulli Distribution:")
print(f"True Probability of Success: {true_prob_bernoulli}, Estimated Probability (MLE): {prob_mle_bernoulli}\n")

print("Exponential Distribution:")
print(f"True Lambda: {true_lambda_exponential}, Estimated Lambda (MLE): {lambda_mle_exponential}")

# Plot the histogram and MLE estimated PDF for the Exponential distribution
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.hist(samples_gaussian, bins=20, density=True, alpha=0.6, color='b', label='Sample Data')
x_gaussian = np.linspace(min(samples_gaussian), max(samples_gaussian), 100)
pdf_estimated_gaussian = norm.pdf(x_gaussian, mu_mle_gaussian, sigma_mle_gaussian)
plt.plot(x_gaussian, pdf_estimated_gaussian, 'r', label='Estimated Gaussian PDF')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Gaussian Distribution')

plt.subplot(132)
plt.hist(samples_bernoulli, bins=[-0.5,0.5,1.5], density=True, alpha=0.6, color='b', label='Sample Data')
plt.axhline(y=true_prob_bernoulli, color='r', linestyle='--', label='Probability of Success')
plt.xticks([0, 1], ['Failure', 'Success'])
plt.legend()
plt.xlabel('Outcome')
plt.ylabel('Probability Density')
plt.title('Bernoulli Distribution')

plt.subplot(133)
plt.hist(samples_exponential, bins=20, density=True, alpha=0.6, color='b', label='Sample Data')
x_exponential = np.linspace(min(samples_exponential), max(samples_exponential), 100)
pdf_estimated_exponential = expon.pdf(x_exponential, scale=1/lambda_mle_exponential)
plt.plot(x_exponential, pdf_estimated_exponential, 'r', label='Estimated Exponential PDF')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution')

plt.tight_layout()
plt.show()
```
